```python
joined = ""
for i in imdb.Overview:
    joined = joined+i+" "
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
plt.figure(figsize = (12,10))
word_cloud = WordCloud(collocations = True, background_color = 'white', stopwords=STOPWORDS).generate(joined)
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.title("Wordcloud showing most frequent words in summary")
plt.show()
```

```python
def number_of_sentences(text):
    for i in text:
        return len(i.split("."))
```

```python
num_sent = []
for i in imdb.Overview:
    num_sent.append(number_of_sentences(i))
```

```python
plt.figure(figsize=(8,6))
sns.countplot(num_sent)
#this shows that all the summaries are in one sentence
plt.title("Plot showing that all the summaries are 1 sentence long")
```

```python
genres = []
for i in imdb.Genre:
    i = i.split(",")
    for j in i:
        genres.append(j.strip())
pd.value_counts(pd.Series(genres))
```

```python
len(pd.value_counts(genres))
```

```python
plt.figure(figsize = (16,8))
sns.countplot(genres)
plt.xticks(rotation = 90, ha = "right")
plt.title("Genres in the dataset")
plt.show()
```

```python
stars = []
for j in [imdb.Star1, imdb.Star2, imdb.Star3, imdb.Star4]:
    for i in j:
        stars.append(i)
```

```python
pd.value_counts(pd.Series(stars))
```

```python
duration = []
for i in imdb.Runtime:
    duration.append(int(i.split(" ")[0]))
```
